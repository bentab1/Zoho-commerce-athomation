```python
import requests

# -------------------------------
# Commerce API Credentials
# -------------------------------
ZOHO_COM_CLIENT_ID = "1000.S0MCEJ7YY6TVO6HC3RED00QLK7KDRA"
ZOHO_COM_CLIENT_SECRET = "1bbb7b39b57c784a38cbdc9c6e82496ee690bde86d"
ZOHO_COM_REFRESH_TOKEN = "1000.e6b8fed22697a4847fa2912eb7bcdc71.db25a2b68ef38e7bf01858ad17489870"
ORG_ID = "891730368"
API_BASE_URL = "https://commerce.zoho.com/store/api/v1"

# -------------------------------
# Helper Functions
# -------------------------------

def get_zoho_commerce_token():
    url = "https://accounts.zoho.com/oauth/v2/token"
    payload = {
        "refresh_token": ZOHO_COM_REFRESH_TOKEN,
        "client_id": ZOHO_COM_CLIENT_ID,
        "client_secret": ZOHO_COM_CLIENT_SECRET,
        "grant_type": "refresh_token"
    }
    resp = requests.post(url, data=payload)
    data = resp.json()
    if "access_token" in data:
        return data["access_token"]
    else:
        raise Exception(f"‚ùå Failed to get Commerce access token: {data}")

def fetch_collections(access_token):
    collections = []
    page = 1
    headers = {"Authorization": f"Zoho-oauthtoken {access_token}",
               "X-com-zoho-store-organizationid": ORG_ID}
    while True:
        url = f"{API_BASE_URL}/collections?page={page}&per_page=200"
        resp = requests.get(url, headers=headers)
        data = resp.json()
        if resp.status_code != 200 or "collections" not in data or not data["collections"]:
            break
        collections.extend(data["collections"])
        if not data.get("page_context", {}).get("has_more_page", False):
            break
        page += 1
    return collections

def choose_collection(access_token):
    collections = fetch_collections(access_token)
    if not collections:
        print("‚ùå No collections found in Zoho Commerce.")
        return None
    print("\nAvailable Collections:")
    for i, c in enumerate(collections, 1):
        print(f"{i}. {c.get('name')} (ID: {c.get('collection_id')})")
    while True:
        choice = input("\nChoose a collection by number or enter Collection ID: ").strip()
        if choice.isdigit() and 1 <= int(choice) <= len(collections):
            return collections[int(choice)-1].get("collection_id")
        else:
            for c in collections:
                if c.get("collection_id") == choice:
                    return choice
            print("‚ùå Invalid choice. Please try again.")

def get_product_ids_from_skus(access_token, skus):
    """Fetch all products and variants, map SKUs to numeric product IDs for collections."""
    product_ids = []
    sku_map = {}
    headers = {
        "Authorization": f"Zoho-oauthtoken {access_token}",
        "X-com-zoho-store-organizationid": ORG_ID
    }

    print("üîç Fetching all products and variants to match SKUs locally...")
    page = 1
    while True:
        url = f"{API_BASE_URL}/products?page={page}&per_page=200"
        resp = requests.get(url, headers=headers)
        try:
            data = resp.json()
        except ValueError:
            print(f"‚ùå Invalid JSON response on page {page}: {resp.text}")
            break

        products = data.get("products", [])
        if not products:
            break

        for p in products:
            # Add main product SKU
            main_sku = p.get("sku")
            pid = p.get("product_id")
            if main_sku and pid:
                sku_map[main_sku.upper()] = pid

            # Add variant SKUs if present
            variants = p.get("variants", [])
            for v in variants:
                v_sku = v.get("sku")
                if v_sku and pid:
                    sku_map[v_sku.upper()] = pid

        if not data.get("page_context", {}).get("has_more_page", False):
            break
        page += 1

    # Match input SKUs
    for sku in skus:
        pid = sku_map.get(sku.upper())
        if pid:
            product_ids.append(pid)
            print(f"‚úÖ Matched SKU {sku} -> Product ID {pid}")
        else:
            print(f"‚ö†Ô∏è SKU not found or invalid: {sku}")

    if not product_ids:
        print("‚ùå No valid product IDs found for given SKUs.")

    return product_ids

def get_collection_items(access_token, collection_id):
    """Fetch all items in a collection with SKU and product name."""
    headers = {"Authorization": f"Zoho-oauthtoken {access_token}",
               "X-com-zoho-store-organizationid": ORG_ID}
    items = []
    page = 1
    while True:
        url = f"{API_BASE_URL}/collections/{collection_id}/products?page={page}&per_page=200"
        resp = requests.get(url, headers=headers)
        data = resp.json()
        if resp.status_code != 200 or "products" not in data or not data["products"]:
            break
        for p in data["products"]:
            sku = p.get("sku")
            name = p.get("name")
            if sku and name:
                items.append({"sku": sku, "name": name})
        if not data.get("page_context", {}).get("has_more_page", False):
            break
        page += 1
    return items

def get_sku_list(access_token, prefilled_skus=None):
    """Prompt user to generate or input SKUs."""
    while True:
        if prefilled_skus:
            print("Available SKUs in collection:")
            for idx, item in enumerate(prefilled_skus, 1):
                print(f"{idx}. {item['sku']} - {item['name']}")
            selection = input("Enter SKU numbers to remove (comma separated) or SKUs directly: ").strip()
            skus = []
            if selection:
                parts = [s.strip() for s in selection.split(",") if s.strip()]
                for p in parts:
                    if p.isdigit() and 1 <= int(p) <= len(prefilled_skus):
                        skus.append(prefilled_skus[int(p)-1]["sku"])
                    else:
                        skus.append(p.upper())
            if skus:
                return None, skus
        else:
            mode = input("Choose SKU input mode:\n1 - Generate SKUs\n2 - Provide SKU list (comma separated)\nEnter choice (1 or 2): ").strip()
            if mode == "1":
                prefix = input("Enter prefix (e.g. AKASKU1): ").strip()
                item_name = input("Enter item name (e.g. IP): ").strip()
                try:
                    start = int(input("Enter start number (e.g. 1): ").strip())
                    end = int(input("Enter end number (e.g. 10): ").strip())
                except ValueError:
                    print("‚ùå Invalid number input.")
                    continue
                mid_code = item_name.upper()[:2]
                pad_option = input("Choose padding type:\n1 - Fixed 6-digit zero padding\n2 - No padding\n3 - Dynamic padding\nEnter choice (1, 2, or 3): ").strip()
                if pad_option == "1":
                    skus = [f"{prefix}-{mid_code}-{str(num).zfill(6)}" for num in range(start, end+1)]
                elif pad_option == "2":
                    skus = [f"{prefix}-{mid_code}-{str(num)}" for num in range(start, end+1)]
                elif pad_option == "3":
                    pad_len = max(len(str(start)), len(str(end)))
                    skus = [f"{prefix}-{mid_code}-{str(num).zfill(pad_len)}" for num in range(start, end+1)]
                else:
                    print("‚ùå Invalid padding option.")
                    continue
            elif mode == "2":
                sku_raw = input("Enter SKUs separated by commas: ").strip()
                skus = [sku.strip().upper() for sku in sku_raw.split(",") if sku.strip()]
                if not skus:
                    print("‚ùå No SKUs entered.")
                    continue
            else:
                print("‚ùå Invalid mode.")
                continue
            return mode, skus
            
def add_products_to_collection(access_token, collection_id, skus):
    product_ids = get_product_ids_from_skus(access_token, skus)
    if not product_ids:
        print("‚ùå No valid products to add.")
        return

    headers = {
        "Authorization": f"Zoho-oauthtoken {access_token}",
        "X-com-zoho-store-organizationid": ORG_ID,
        "Content-Type": "application/json"
    }
    payload = {"product_ids": product_ids}  # Must be numeric IDs
    resp = requests.post(f"{API_BASE_URL}/collections/{collection_id}/products", json=payload, headers=headers)
    data = resp.json()
    if data.get("code") == 0:
        print(f"‚úÖ Added {len(product_ids)} products successfully.")
    else:
        print("‚ùå Failed to add products:", data)

def remove_products_from_collection(access_token, collection_id, skus):
    headers = {"Authorization": f"Zoho-oauthtoken {access_token}",
               "X-com-zoho-store-organizationid": ORG_ID,
               "Content-Type": "application/json"}
    payload = {"product_skus": skus}
    resp = requests.delete(f"{API_BASE_URL}/collections/{collection_id}/products", json=payload, headers=headers)
    data = resp.json()
    if data.get("code") == 0:
        print(f"‚úÖ Successfully removed {len(skus)} SKUs from collection {collection_id}")
    else:
        print("‚ùå Failed to remove products:", data)

# -------------------------------
# Main Script
# -------------------------------

def main():
    access_token = get_zoho_commerce_token()

    print("Select action:")
    print("1 - Add item(s) to collection")
    print("2 - Delete item(s) from collection")
    choice = input("Enter choice (1 or 2): ").strip()

    collection_id = choose_collection(access_token)
    if not collection_id:
        return

    if choice == "1":
        _, skus = get_sku_list(access_token)
        if skus:
            add_products_to_collection(access_token, collection_id, skus)
    elif choice == "2":
        items_in_collection = get_collection_items(access_token, collection_id)
        if not items_in_collection:
            print("‚ùå Collection is empty.")
            return
        _, skus = get_sku_list(access_token, prefilled_skus=items_in_collection)
        if skus:
            remove_products_from_collection(access_token, collection_id, skus)
    else:
        print("‚ùå Invalid choice.")

if __name__ == "__main__":
    main()
```
